```python
import numpy as np
#1 Suma de Matrices rectangulares

A = np.array([[1,1,1],[2,1,2]]) #Se declaran 2 matrices como ejemplo
B = np.array([[1,1,1],[1,1,1]])

columnas = len(A[0,:]) #Se ve la cantidad de columnas que tienen las matrices
filas = len(A[:,0]) #Se ve la cantidad de filas de las matrices
#Hacemos una función que haga la suma
def suma_matrices(A,B):
  S = np.zeros(shape=(filas,columnas))
  """Declaramos una matriz S de ceros donde se va a almacenar la suma
  del mismo tamaño que la matriz A"""
  for i in range(filas):
     for j in range(columnas):
      S[i,j] = A[i,j]+ B[i,j]
      """En base a la longitud de las filas y columnas,
      hacemos que para cada elemento ij en las matrices A y B, se relice su suma
      y se almacene en la misma posición ij, en la matriz S"""
  return(S)

suma_matrices(A,B)
```

```python
import numpy as np
#2 Multiplicación de matrices entrada a entrada

A = np.array([[1,1,1],[2,1,2]]) #Declaramos las matrices
B = np.array([[1,1,1],[1,1,1]])

columnas = len(A[0,:])
filas = len(A[:,0])

def prod_matrices(A,B): #definimos una función que realize el produco entrada a entrada
  P = np.zeros(shape=(filas,columnas))
  """Declaramos una matriz P de ceros donde se va a almacenar la suma
  del mismo tamaño que la matriz A"""
  for i in range (filas):
     for j in range(columnas):
      P[i:j] = A[i:j]*B[i:j]
      """En base a la longitud de las filas y columnas,
      hacemos que para cada elemento ij en la matriz, A relice el producto
      con su correspondiente ij en la Matriz B y se almacene en la misma posición (ij), en la matriz P"""
  return(P)

prod_matrices(A,B)
```

```python
import numpy as np
#3 Traza de una matriz
A = np.array([[1,1,1],[2,2,2],[3,3,3]])
#Se declara con una matriz, tiene que ser cuadrada para realizar la traza
def traza (A): #definimos nuestra función traza
  tr = 0 #le damos un valor inicial de 0 a la traza
  for i in range(len(A[0,:])):
    tr = A[i,i]+tr #creamos un bucle donde la traza se va ir sumando con los elementos de la for a_ii
  traza = "Traza = "+str(tr) #Pasamos a string nuestra traza
  return(traza) #le pedimos nos regrese la string de traza

traza(A) #realizamos la traza
```

```python
import numpy as np
# 4 transpuesta de una matriz
A = np.array([[1,1,1],[2,2,2],[3,3,3]]) #declaramos una matriz como ejemplo
columnas = len(A[0,:]) #vemos la cantidad de columnas que tiene
filas = len(A[:,0]) #vemos la cantidad de filas que tiene
Transpuesta = np.zeros(shape=(filas,columnas)) #Declaramos una matriz Transpuesta,
#que este formada por solo ceros, y que sea del mismo tamaño que la matriz A
def trans(A):#hacemos una función para hacer la transpuesta
  for i in range(columnas): #pedimos que para cada "i" columna
#ahora de vuelva la "i" fila de la transpuesta
    Transpuesta[:,i] = A[i,:]
  for j in range(filas):
    Transpuesta[j,:] = A[:,j]
#pedimos que para cada "j" fila
#ahora de vuelva la "j" columna de la transpuesta
  return(Transpuesta)
trans(A) #Hacemos la transpuesta de A como ejemplo
```

```python
import numpy as np
import matplotlib.pyplot as plt

def y (x, C):
    return (x - 1) + C * np.exp(-x) # definimos la función y(x, C) = (x - 1) + C * e^(-x)

x = np.linspace(-2, 5, 500) # creamos un rango de valores para x

# definimos los valores de C para los tres casos
C_positive = [1, 2, 3]  # C > 0
C_zero = [0]            # C = 0
C_negative = [-1, -2, -3]  # C < 0

#plt.figure(figsize=(10, 6))

# f'C = {C}' crea etiquetas dinámicas para las curvas graficadas, reflejando el valor actual de 𝐶 en cada una de ellas.
for C in C_positive:
    plt.plot(x, y(x, C), alpha = 0.85, label = f'C ={C}' , ls = '--')  # usamos label para nombrar cada línea

for C in C_zero:
    plt.plot(x, y(x, C), label = f'C = {C}', color = 'black', linewidth = 2)  # usamos color para cambiar el color de la línea

for C in C_negative:
    plt.plot(x, y(x, C), alpha = 0.85, label = f'C = {C}', ls = '-.')  # usamos ls para cambiar el tipo de línea

plt.title('Familia de curvas', fontsize = 14)
plt.xlabel('x', fontsize = 12)
plt.ylabel('y(x)', fontsize = 12)

plt.grid(ls = ':', alpha = 0.7)

plt.legend(title = "Valores de C")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (3 * x + 7) / (x**2 - 4) # definimos la función: f(x) = (3x + 7) / (x^2 - 4)

# creamos un rango de valores para x, excluyendo los puntos donde el denominador es cero (x = ±2)
x = np.linspace(-10, 10, 1001)
x = x[x != -2]  # Excluimos x = -2 (asíntota vertical)
x = x[x != 2]   # Excluimos x = 2 (asíntota vertical)

# identificamos las asíntotas (verticales y horizontal)
horizontal_asymptote = 0  # Coeficientes principales: 3/inf (0)
vertical_asymptotes = [-2, 2]

plt.plot(x, f(x), "-b", alpha = 0.6, label = 'f(x) = (3x + 7) / (x^2 - 4)') # graficamos la función

for va in vertical_asymptotes:
    plt.axvline(va, color = 'r', ls = '--', label = f'Asíntota vertical: x = {va}') # añadimos las asíntotas verticales (líneas punteadas color rojo)

plt.axhline(horizontal_asymptote, color = 'g', ls = '--', label = 'Asíntota horizontal: y = 0') # añadimos la asíntota horizontal (línea punteada color verde)

# configuración del eje y límites
plt.ylim(-10, 10)  # ajustamos para evitar valores extremos
plt.xlim(-10, 10)

plt.title('Gráfica de f(x)', fontsize = 14)
plt.xlabel('x', fontsize = 12)
plt.ylabel('f(x)', fontsize = 12)

plt.legend()

plt.grid(alpha = 0.7, ls = ':')

plt.show()
```
